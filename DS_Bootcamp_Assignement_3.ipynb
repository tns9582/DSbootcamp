2.Track pedestrian counts on the Brooklyn Bridge for the year 2019 and analyze how different weather conditions influence pedestrian activity in that year:


```python
import pandas as pd

# Read the dataset
url = "https://data.cityofnewyork.us/api/views/6fi9-q3ta/rows.csv?accessType=DOWNLOAD"
df = pd.read_csv(url)

# Convert 'hour_beginning' column to datetime format
df['hour_beginning'] = pd.to_datetime(df['hour_beginning'])

# Filter data for the year 2019 and Brooklyn Bridge location
brooklyn_bridge_2019 = df[(df['hour_beginning'].dt.year == 2019) & (df['location'] == 'Brooklyn Bridge')]

# Analyze pedestrian counts by weather summary
pedestrian_counts_by_weather = brooklyn_bridge_2019.groupby('weather_summary')['Pedestrians'].sum()

# Print pedestrian counts by weather summary
print("Pedestrian counts by weather summary:")
print(pedestrian_counts_by_weather)

# Analyze the correlation between weather summary and pedestrian counts
correlation_matrix = brooklyn_bridge_2019[['Pedestrians', 'weather_summary']].corr()

# Print correlation matrix
print("\nCorrelation Matrix:")
print(correlation_matrix)
```

3. Implement a custom function to categorize time of day into morning, afternoon, evening, and night:


```python
import pandas as pd

# Read the dataset
url = "https://data.cityofnewyork.us/api/views/6fi9-q3ta/rows.csv?accessType=DOWNLOAD"
df = pd.read_csv(url)

# Convert 'hour_beginning' column to datetime format
df['hour_beginning'] = pd.to_datetime(df['hour_beginning'])

# Define a function to categorize time of day
def categorize_time_of_day(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 20:
        return 'Evening'
    else:
        return 'Night'

# Apply the function to create a new column for time of day categories
df['Time of Day'] = df['hour_beginning'].dt.hour.apply(categorize_time_of_day)

# Print the DataFrame to verify the new column
print(df[['hour_beginning', 'Time of Day']])

```
